# 删除重复值

```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({'k1':['one', 'two']*3+['two'],
                    'k2':[1, 1, 2, 3, 3, 4, 4]})
```

```python
data
```

```python
data.duplicated()
```

```python
data.drop_duplicates()
```

```python
data['v1'] = range(7)
```

```python
data
```

```python
data.drop_duplicates(['k1'])
```

```python
data.duplicated('k1')
```

```python
data.drop_duplicates(['k1', 'k2'])
```

```python
data.duplicated(['k1','k2'])
```

## 联合重叠数据

```python
import pandas as pd
import numpy as np
```

```python
a = pd.Series([np.nan, 2.5, 0.0, 3.5, 4.5, np.nan], 
              index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series([0., np.nan, 2., np.nan, np.nan, 5.0], 
              index=['a', 'b', 'c', 'd', 'e', 'f'])
```

```python
a
```

```python
b
```

```python
pd.isnull(a)
```

```python
np.where(pd.isnull(a), b, a)
```

```python
b.combine_first(a)
```

## DataFrame

```python
df1 = pd.DataFrame({'a': [1, np.nan, 5, np.nan],
                   'b': [np.nan, 2, np.nan, 6],
                   'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a': [5, 4, np.nan, 5, 7],
                   'b': [np.nan, 3, 4, 6, 8]})
```

```python
df1
```

```python
df2
```

```python
df1.combine_first(df2)
```
